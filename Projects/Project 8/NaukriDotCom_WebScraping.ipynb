### Problem Statement : Scrape The Data From JobStreet And Naukri.com. 
Save the jobs info in csv files. You will create one csv file for JobStreet and one for Naukri.com.

```python
# Import Necessary libraries
import pandas as pd
import numpy as np
import time
import requests
import selenium
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from bs4 import BeautifulSoup


# Generating the URL
def get_url(position,location):
    position = position.replace(' ',"-")
    location = location.replace(' ',"-")
    mid_text = position+"-jobs-in-"+location
    temp = "https://www.naukri.com/"+mid_text+"?k={}&l={}"  #https://naukri.com/python-jobs-in-pune?k=python&l=pune
    url = temp.format(position,location)
    return url

url = get_url('java dev', 'pune')
driver = webdriver.Firefox()
driver.get(url)
time.sleep(10)

soup = BeautifulSoup(driver.page_source,'html.parser') 


def find_job(card):
    # Scraping job_name
    try:
        job_name = card.find('div','row1').text
    except:
        job_name = 'Not Mentioned'
        
    # Scraping company_name
    try:
        company_name = card.find('span','comp-dtls-wrap').find('a','comp-name').text
    except:
        company_name = 'Not Mentioned'
        
    # Scraping company_location
    try:
        company_location = card.find('span','locWdth').text
    except:
        company_location = 'Not Mentioned'
        
    # Scraping salary
    try:
        salary = card.find('span','sal-wrap ver-line').text
    except:
        salary = 'Not Mentioned'
        
    # Scraping job_description
    try:
        job_description = card.find('span','job-desc ni-job-tuple-icon ni-job-tuple-icon-srp-description').text
    except:
        job_description = 'Not Mentioned'
        
    # Scraping job_apply_url
    try:
        job_apply_url = card.find('div','row1').find('a')
        job_apply_url = job_apply_url['href']
    except:
        job_apply_url = 'Not Mentioned'


    # Scraping Job Posting Date
    try:
        job_posted = card.find('span','job-post-day').text
    except:
        job_posted = 'Not Mentioned'
        
    all_job_detail = (job_name,company_name,company_location,salary,job_description,job_posted,job_apply_url)
    return all_job_detail


# main driver function
def main():
    all_jobs = []
    
    cards = soup.find_all('div','srp-jobtuple-wrapper')
    for i in cards:
        temp = find_job(i)
        all_jobs.append(temp)
    col = ['Job Name','Company','Company Location','Salary','Job Description','Job Posted','Apply url']
    naukri_jobs_data = pd.DataFrame(all_jobs,columns = col)
    naukri_jobs_data.to_csv(r'C:\Users\Harshan Attar\Desktop\iGurus Practice\jobs\NaukriDotCom.csv',encoding='utf-8')
```

```python
main()
```
